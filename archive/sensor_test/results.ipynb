```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal

GAIN_VALS = [10, 20, 30, 40, 50]
LOW_CUTOFF = 10
HIGH_CUTOFF = 450
SAMPLE_RATE = 1152
```

```python
# Open the .txt containing the sensor output
# and then store the values for plotting
raw_data = []
for i in GAIN_VALS:
  with open(f"gain{i}.txt", "r") as f:
    temp = f.read().splitlines()
    temp = temp[:30000]
    num_points = len(temp)
    temp_data = np.zeros(num_points)
    for c, t in enumerate(temp):
      temp_data[c] = t
    raw_data.append(temp_data)
```

```python
fig, axs = plt.subplots(len(raw_data), sharex=True)
fig.suptitle('Comparing gain values')
fig.tight_layout()

for c, i in enumerate(raw_data):
  axs[c].plot(i)
  axs[c].set_title(f"Gain: {c}")

plt.show()
```

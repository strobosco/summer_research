# Single Arm Data Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open the .txt containing the sensor output
# and then store the values

FILENAME = "arm-test.txt"
with open(FILENAME, "r") as f:
  temp = f.read().rstrip().split()

data = np.array([int(i) for i in temp])
print(data)
```

```python
# 50 milliseconds delay in between samples, so create time axis for plot
time_axis = np.array([i/1000 for i in range(0, 50 * len(data), 50)])

plt.plot(time_axis, data)
```

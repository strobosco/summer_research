# Single Arm Data Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Accessing on log at a time

```python
# Open the .txt containing the sensor output
# and then store the values

FILENAME = "rock.txt"

with open(FILENAME, "r") as f:
  temp = f.read().splitlines()

num_points = len(temp)
data = np.zeros((3, num_points))
for c, t in enumerate(temp):
  split = t.split(',')
  data[0][c] = split[0]
  data[1][c] = split[1]
  data[2][c] = split[2]
  
print(len(data[0]))
```

```python
# 50 milliseconds delay in between samples, so create time axis for plot
time_axis = np.array([i/1000 for i in range(0, 50 * num_points, 50)])

plt.plot(time_axis, data[0])
plt.plot(time_axis, data[1])
plt.plot(time_axis, data[2])
plt.title("Rock")

print(len(data[0]))
print(len(time_axis))
```

# Accessing all logs

```python
FILENAMES = ["rock.txt", "paper.txt", "scissors.txt"]
names = {0 : "Rock", 1 : "Paper", 2 : "Scissors"}

stored_logs = []

for file in FILENAMES:
  with open(file, "r") as f:
    temp = f.read().splitlines()

  num_points = len(temp)
  data = np.zeros((3, num_points))
  for c, t in enumerate(temp):
    split = t.split(',')
    data[0][c] = split[0]
    data[1][c] = split[1]
    data[2][c] = split[2]
  stored_logs.append(data)
```

```python
# 50 milliseconds delay in between samples, so create time axis for plot
fig, axs = plt.subplots(len(stored_logs))
fig.suptitle('EMG data for diffefent gestures')

for c, i in enumerate(stored_logs):
  time_axis = np.array([i/1000 for i in range(0, 50 * len(i[0]), 50)])
  axs[c].plot(time_axis, i[0])
  axs[c].plot(time_axis, i[1])
  axs[c].plot(time_axis, i[2])
  axs[c].set_title(names.get(c))

plt.show()
```
